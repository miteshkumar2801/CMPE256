```python
#5 Betweenness Centrality
betCen=nx.betweenness_centrality(g)
```

```python
betCensorted = list(betCen.values())
betCensorted.sort(reverse=True)
betCentop5= betCensorted[0:5]
top5Nodes = []

for key in betCen:
    if betCen[key] in betCentop5:
        top5Nodes.append(key)


for i in top5Nodes:
    print (i, betCen[i])
```

```python
def top5fromDict(dict):
    
    #print (list(dict.values())).sort(reverse=True)
    valuesSorted = list(dict.values())
    valuesSorted.sort()
    top5val = valuesSorted[0:5]
    top5keys = []
    
    top5keys = []
    count = 0
    for i in range(0,5):
        for key in dict:
            if dict[key] == top5val[i]:
                top5keys.append(key)
                #print(count)
                count += 1
                if count == 5:
                    return top5keys
                
                
    return top5keys
                    
    
    
```

```python
## Top 5 pageRank
top5Nodes = top5fromDict(pgRank)
print (top5Nodes)
```

```python
#7 Hits
hubs,auth = nx.hits(g)
top5auth = top5fromDict(auth)
print (top5auth)
```
